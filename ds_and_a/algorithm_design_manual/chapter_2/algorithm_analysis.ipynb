**RAM model of computation** - what he means by this is an ideal computer that is useful in estimating runtime of algorithms. Each operation (+, -, *, if, call) takes one step. Loops and subroutines are composed of many simple operations. Each memory access is one step, regardless of whether something is in cache or on the disk.

The RAM model does not describe how computers actually work. Multiplication takes more than one step. Compilers can unroll loops. Hyperthreading exists. Memory-access times vary greatly depending on where data sits in the storage heirarchy.

Nevertheless, RAM provides an excellent model for algorithm performance.

# Big ùòñ notation

when using Big ùòñ notation, think of the `==` as meaning "one of the functions that are".

## ùòñ is the upper bound
`f(n) == ùòñ(g(n))` means `c ‚àô g(n)` is an upper bound on `f(n)`. Thus, there exists some constant `c` such that `f(n) <= c ‚àô g(n)` for every large enough n.


## ùõ∫ is the lower bound
`f(n) = ùõ∫(g(n))` means `c ‚àô g(n)` is a lower bound on f(n). Thus, there exists some constant `c` such that `f(n) >= c ‚àô g(n)` for all `n >= n0`


## ùõ≥ is both upper and lower bounds
`f(n) == ùõ≥(g(n))` means there exist c1 and c2 such that `f(n) <= c1 ‚àô g(n)` and `f(n) <= c2 ‚àô g(n)` for all `n>=n0`. Essentially, `g(n)` provides a nice tight bound on `f(n)`. If both `ùòñ(g(n))` and `ùõ∫(g(n))`, then `ùõ≥(g(n))`.

## Classes of functions commonly discussed with Big O notation
n is the number of elements
c is some arbitrary constant

Listed from slowest growing to fastest growing

| notation                |class|
|-------------------------|-----|
| O(n)                    |constant|
| O(log(n))               |logarithmic|
| O((log(n))<sup>c</sup>) |polylogarithmic|
| O(n)                    |linear|
| O(n<sup>2</sup>)        |quadratic|
| O(n<sup>c</sup>)        |polynomial|
| O(c<sup>n</sup>)        |exponential|

```python
import matplotlib.pyplot as plt
from numpy import ndarray
import math

funcs = [
    (lambda x: 3, 'constant'),
    (lambda x: math.log(x), 'logarithmic'),
    (lambda x: x, 'linear'),
    (lambda x: x**2, 'quadratic'),
    (lambda x: x**3, 'polynomial'),
    (lambda x: 2**x, 'exponential')
]

last_x = 100

num_rows = len(funcs)//2 if len(funcs) % 2 == 0 else len(funcs)//2 + 1
num_columns = 2

fig, axs = plt.subplots(num_rows, num_columns, figsize = (9,9), dpi=120)
xs = list(range(1, last_x))

for func, ax in zip(funcs, ndarray.flatten(axs)):
    ys = [func[0](x) for x in xs]
    ax.plot(xs, ys, 'b')
    ax.set_title(func[1])

fig.subplots_adjust(hspace=.5, wspace=0.5)
plt.show()
```

_Problem_: is (x + y)<sup>2</sup> == ùòñ(x<sup>2</sup> + y<sup>2</sup>)?
_Solution_: Working with Big O means going back to the definition at the slightest sign of confusion. By definition, this is expression is valid iff we can find some `c` such that (x + y)<sup>2</sup> <= ùòñ(x<sup>2</sup> + y<sup>2</sup>) for all sufficiently large x and y.
1. expand left side -> x<sup>2</sup> + 2xy + y<sup>2</sup>
2. if x <= y, then 2xy <= 2y<sup>2</sup> <= 2(x<sup>2</sup> + y<sup>2</sup>).
3. if x >= y, then 2xy <= 2x<sup>2</sup> <= 2(x<sup>2</sup> + y<sup>2</sup>).
4. Either way, we can bound the 2xy by two times the right-side function x<sup>2</sup> + y<sup>2</sup>.
5. This means that (x + y)<sup>2</sup> <= 3(x<sup>2</sup> + y<sup>2</sup>), and so the result holds.
